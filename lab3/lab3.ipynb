## zadanie 1

```python
def lagrange(xp, X, Y):
    yp = 0
    n = len(X)
    for i in range(n):
        p = 1
        for j in range(n):
            if i != j:
                p = p * (xp - X[j])/(X[i] - X[j])
    
        yp = yp + p * Y[i]  
        
    return yp
```

```python
def f(x):
    return 1/(1 + x**2)
```

```python
N = [5, 10, 15]
```

```python
for n in N:
    ax = np.linspace(-5, 5, 200)
    X = np.linspace(-5, 5, n)
    Y = [f(x) for x in X]
    y_lagrange = [lagrange(x, X, Y) for x in ax]
    y_f = [f(x) for x in ax]
    plt.plot(ax, y_f)
    plt.plot(ax, y_lagrange)
    x_err = np.linspace(-5, 5, 30)
    y_err = [f(x) - lagrange(x, X, Y) for x in x_err]
    plt.plot(x_err, y_err)
    plt.show()
```
